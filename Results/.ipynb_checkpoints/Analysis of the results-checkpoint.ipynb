# Analysis of the results

En el siguiente experimento se ha realizado un embedding aplicando word2vec con los siguientes hiperparámetros:

size (dimensión en la que se embebe) = 150;
window = 11;
nº de iteraciones = 250.

A continuación se realizó la filtración Vietoris-Rips para dimensiones 0 y 1 con métrica euclídea y coseno. 

Finalmente, en este documento se estudian los resultados obtenidos después de 100 repeticiones del experimento. La conclusión es que la distancia coseno para dimensión 0 es capaz de distinguir y establece que Quevedo y lope están cerca. Sin embargo, la distancia Euclidea establece que Quevedo y Góngora son los más cercanos.

```python
import numpy as np
```

## Cosine distance

### Dim 0

```python
gongora_lope_cos = np.loadtxt("2dic_lopevsgongora_dim0_cos")
quevedo_gongora_cos = np.loadtxt("2dic_quevedovsgongora_dim0_cos")
quevedo_lope_cos = np.loadtxt("2dic_quevedovslope_dim0_cos")
```

```python
import matplotlib.pyplot as plt
plt.scatter( range(0,100),gongora_lope_cos,label='Gongora vs Lope')#,marker='.')
plt.scatter( range(0,100),quevedo_gongora_cos,label='Gongora vs Quevedo')#,marker='.')
plt.scatter( range(0,100),quevedo_lope_cos,label='Quevedo vs Lope',)#marker='.')
plt.legend(title='Poets comparison where:')
plt.show()
```

```python
print("Media distancia euclidea Quevedo y Lope:",np.mean(quevedo_lope_cos))
print("Media distancia euclidea Quevedo y Gongora:",np.mean(quevedo_gongora_cos))
print("Media distancia euclidea Gongora y Lope:",np.mean(gongora_lope_cos))
```

```python
data = [quevedo_lope_cos, quevedo_gongora_cos, gongora_lope_cos]
fig7, ax7 = plt.subplots()
ax7.set_title('Poets comparison')
ax7.boxplot(data)

plt.show()
```

```python
len(quevedo_lope_cos)
```

```python
import numpy as np
from scipy import stats

#Sample Size
N = 100
a = quevedo_lope_cos
b = quevedo_gongora_cos


## Cross Checking with the internal scipy function
t2, p2 = stats.ttest_ind(b,a)
print("t = " + str(t2))
print("p = " + str(p2))
```

```python
import numpy as np
from scipy import stats

#Sample Size
N = 100
a = quevedo_lope_cos
b = gongora_lope_cos


## Cross Checking with the internal scipy function
t2, p2 = stats.ttest_ind(b,a)
print("t = " + str(t2))
print("p = " + str(p2))
```

Quevedo y Lope serían los más cercanos y distingue bien. Los outliers pueden considerarse entrenamientos fallidos.
